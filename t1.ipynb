```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
```

```python
x1 = np.random.rand(5, 1) * 20
y1 = 2 * x1 - 3
y1 = y1 + np.random.normal(scale=1, size = y1.shape)
x1 = x1.reshape(-1,1)
y1 = y1.reshape(-1,1)
```

```python
lr = LinearRegression(normalize=True)
lr.fit(x1,y1)
print(lr.coef_,lr.intercept_)
```

```python
h = lr.predict(x1)
MSE = np.mean((y1 - h) ** 2)/2
print(MSE)
```

```python
lr.score(x1,y1)
```

```python
X1 = np.arange(0,25,0.1).reshape(-1,1)
Y1 = lr.predict(X1)

plt.plot(x1,y1,"bo")
plt.plot(X1,Y1,"r--")
plt.xlabel('X -axe')
plt.ylabel('Y -axe')
plt.title('5 data and pow = 1')
```

```python
poly = PolynomialFeatures(degree=4, include_bias=False)
x1_p4 = poly.fit_transform(x1)
```

```python
lr1_4 = LinearRegression(normalize=True).fit(x1_p4, y1)
X1_4 = np.arange(-1,22,0.1).reshape(-1,1)
X1_line_P4 = poly.fit_transform(X1_4)
Y1_p4 = lr1_4.predict(X1_line_P4)

plt.plot(x1,y1,"bo")
plt.plot(X1_4,Y1_p4,"r--")
plt.title('5 data and pow = 4')
```

```python
lr1_4.coef_
```

```python
lr1_4.score(x1_p4, y1)
```

```python
poly2 = PolynomialFeatures(degree=16, include_bias=False)
x1_p16 = poly2.fit_transform(x1)
```

```python
lr1_16 = LinearRegression(normalize=True).fit(x1_p16, y1)
X1_16 = np.arange(0,21,0.1).reshape(-1,1)
X1_line_P16 = PolynomialFeatures(degree=16, include_bias=False).fit_transform(X1_16)
Y1_p16 = lr1_16.predict(X1_line_P16)

plt.plot(x1,y1,"bo")
plt.plot(X1_16,Y1_p16,"r--")
plt.title('5 data and pow = 16')
```

```python
lr1_16.coef_
```

```python
lr1_16.score(x1_p16, y1)
```
